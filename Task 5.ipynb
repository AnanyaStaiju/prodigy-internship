```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a sample dataset
data = {
    'Start_Time': pd.date_range(start='2023-01-01', periods=100, freq='H'),
    'Weather_Condition': np.random.choice(['Clear', 'Cloudy', 'Rainy', 'Snowy', 'Foggy'], 100),
    'Severity': np.random.choice([1, 2, 3, 4], 100),  # Severity from 1 to 4
    'Start_Lat': np.random.uniform(low=35.0, high=40.0, size=100),  # Random latitude values
    'Start_Lng': np.random.uniform(low=-120.0, high=-115.0, size=100)  # Random longitude values
}

# Convert to DataFrame
traffic_data = pd.DataFrame(data)

# Display the first few rows of the dataset
print("Sample Traffic Accident Dataset:")
display(traffic_data.head())  # Using display for better visualization in Jupyter

# Visualize accidents by weather condition
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Condition', data=traffic_data, order=traffic_data['Weather_Condition'].value_counts().index, palette='Blues')
plt.xticks(rotation=90)
plt.title('Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()

# Visualize accidents by severity
plt.figure(figsize=(10, 6))
sns.countplot(x='Severity', data=traffic_data, palette='Reds')
plt.title('Accidents by Severity')
plt.xlabel('Severity Level')
plt.ylabel('Number of Accidents')
plt.show()

# Visualize accidents by time of day
traffic_data['Hour'] = traffic_data['Start_Time'].dt.hour

plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=traffic_data, palette='coolwarm')
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()

# Scatter plot for accidents based on Latitude and Longitude
plt.figure(figsize=(10, 8))
plt.scatter(traffic_data['Start_Lng'], traffic_data['Start_Lat'], c=traffic_data['Severity'], cmap='viridis', s=50, alpha=0.7)
plt.colorbar(label='Severity')
plt.title('Accidents Location Scatter Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a sample traffic accident dataset
data = {
    'Start_Time': pd.date_range(start='2023-01-01', periods=200, freq='H'),
    'Weather_Condition': np.random.choice(['Clear', 'Cloudy', 'Rainy', 'Snowy', 'Foggy'], 200),
    'Road_Condition': np.random.choice(['Dry', 'Wet', 'Icy', 'Snowy'], 200),
    'Severity': np.random.choice([1, 2, 3, 4], 200),
    'Start_Lat': np.random.uniform(low=35.0, high=40.0, size=200),
    'Start_Lng': np.random.uniform(low=-120.0, high=-115.0, size=200)
}

# Convert to a DataFrame
traffic_data = pd.DataFrame(data)

# Display the first few rows of the dataset
print("Sample Traffic Accident Dataset:")
display(traffic_data.head())

# Analyze accidents by Weather Condition
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Condition', data=traffic_data, order=traffic_data['Weather_Condition'].value_counts().index, palette='Blues')
plt.xticks(rotation=90)
plt.title('Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()

# Analyze accidents by Road Condition
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Condition', data=traffic_data, order=traffic_data['Road_Condition'].value_counts().index, palette='Greens')
plt.xticks(rotation=90)
plt.title('Accidents by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.show()

# Analyze accidents by Severity
plt.figure(figsize=(10, 6))
sns.countplot(x='Severity', data=traffic_data, palette='Reds')
plt.title('Accidents by Severity')
plt.xlabel('Severity Level')
plt.ylabel('Number of Accidents')
plt.show()

# Analyze accidents by Hour of Day
traffic_data['Hour'] = traffic_data['Start_Time'].dt.hour
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=traffic_data, palette='coolwarm')
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()

# Scatter plot for accidents based on Latitude and Longitude
plt.figure(figsize=(10, 8))
plt.scatter(traffic_data['Start_Lng'], traffic_data['Start_Lat'], c=traffic_data['Severity'], cmap='viridis', s=50, alpha=0.7)
plt.colorbar(label='Severity')
plt.title('Accidents Location Scatter Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

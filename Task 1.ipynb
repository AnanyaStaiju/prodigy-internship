```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_data = pd.read_csv("C:/Users/anany/Downloads/titanic.csv")

# Display the first few rows of the dataset
print(titanic_data.head())

# 1. Bar chart for distribution of gender (categorical variable)
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=titanic_data, palette='coolwarm')
plt.title('Distribution of Gender in the Titanic Dataset')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# 2. Histogram for distribution of age (continuous variable)
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribution of Age in the Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
